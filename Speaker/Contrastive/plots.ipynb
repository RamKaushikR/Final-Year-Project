```python
import pickle
import matplotlib.pyplot as plt
```

```python
lstm = pickle.load(open('Model/LSTM/lstm_history.pickle', 'rb'))

plt.plot(lstm['loss'])
plt.plot(lstm['val_loss'])
plt.title('LSTM Loss')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend(['train', 'val'], loc='upper right')
plt.savefig('Plots/Contrastive_LSTM.png')
```
